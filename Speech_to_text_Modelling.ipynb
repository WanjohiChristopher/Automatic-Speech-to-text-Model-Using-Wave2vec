```python
import torch
import librosa
import numpy as np
import soundfile as sf
from scipy.io import wavfile
from IPython.display import Audio
from transformers import Wav2Vec2ForCTC, Wav2Vec2Tokenizer
from google.colab import drive
drive.mount('/content/drive')
```

```python
#downloading the pretrained model
tokenizer = Wav2Vec2Tokenizer.from_pretrained("facebook/wav2vec2-base-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")
```

```python
#the filename to use
file_name = '/content/OSR_us_000_0038_8k.wav'
Audio(file_name)
```

```python
#the filename to use
filename = '/content/drive/MyDrive/SpeechRecognitionFiles/Welcome.wav'
Audio(filename)
```

```python
#reading the filename
data = wavfile.read(filename)

framerate = data[0]
sounddata = data[1]
time = np.arange(0,len(sounddata))/framerate
print('Sampling rate:',framerate,'Hz')
```

```python
#loading the audio using librosa

input_audio, _ = librosa.load(filename, sr=16000)
```

```python
# Taking an input values
input_values = tokenizer(input_audio, return_tensors="pt").input_values
# Storing logits (non-normalized prediction values)
logits = model(input_values).logits
#Passing the logit values to softmax to get the predicted values using pytorch
predicted_ids = torch.argmax(logits, dim=-1)

#Step to convert audio to text
## Passing the prediction to the tokenizer decode to get the transcription
transcription = tokenizer.batch_decode(predicted_ids)[0]
#printing transcription text
print(transcription)
```
